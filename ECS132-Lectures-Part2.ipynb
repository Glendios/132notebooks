# III. Formal Definition of Probability 
Consider the probability space consisting of $P$ and $S$ where $S$ is the sample space and $P$ is a function that takes as an input an event A and returns $P(A) \in [0,1]$ such that  
1. [Axiom 1:] $P(\Phi) = 0$
2. [Axiom 2:] $P(S) = 1$,
3. [Axiom 3:] If $A_1, A_2, \ldots,$ are disjoint non-overlapping events then
\begin{equation*}
        P(\bigcup_{i=1}^{\infty} A_i) = \sum_{i=1}^{\infty} P(A_i)
\end{equation*}

![Union-of-non-overlapping-events](./Figures/Union-of-non-overlapping-events.jpg)


### Properties

Based on the above axioms we can derive the following properties

1. Property 1: If $A^\complement$ denotes the complement of event $A$, then $P(A^\complement) = 1 - P(A)$.  

2. Property 2: If event $A$ is contained in event $B$, i.e., $A \subseteq B$, then $P(A) \leq P(B)$  

3. Property 3: Consider two events $A, B$, then  
\begin{equation*}
      P(A \cup B) = P(A) + P(B) - P(A \cap B)
\end{equation*}
All of the above properties can be proven from the axioms.  Below we show the proof of Property 2. The proofs of the other two properties are left as exercises.  


#### Proof of Property 2

Prove that if event $A$ is contained in event $B$, i.e., $A \subseteq B$, then $P(A) \leq P(B)$.  

![Two events](./Figures/Two-events.jpg)


We divide $B$ as a union of two mutually exclusive parts $A$ and $(A^\complement \cap B)$. This is best seen from the figure above.   


Thus,
\begin{equation*}
  B = A \cup (A^\complement \cap B)
\end{equation*}
which implies that
\begin{equation*}
 P(B) = P(A) +  P(A^\complement \cap B)
\end{equation*}
since $P(A^\complement \cap B) \geq 0$, $P(B) \geq  P(A)$


### Set theory:

We will also use **deMorgan's laws**

1) $\overline{(A \cup B)} = \overline{A} \cap \overline{B}$

2) $\overline{(A \cap B)} = \overline{A} \cup \overline{B}$

And the **distributive law**.  Here is [a review of the fundamental properties of set algebra.](https://en.wikipedia.org/wiki/Algebra_of_sets)


## III.A Inclusion-Exclusion Principle


First let us consider three events $A_1, A_2$, and $A_3$ as shown in  the figure below. 

![Three Events](./Figures/three-events-A-B-C.jpg)

*Three events A1, A2, and A3. The different intersections are shown. Note the area which is the intersection of all three events A1, A2, and A3.*

\begin{eqnarray}
P(A_1 \cup A_2 \cup A_3) &=&  P(A_1) + P(A_2) + P(A_3) \\
						 & &  - P(A_1 \cap A_2) - P(A_1 \cap A_3) - P(A_2 \cap A_3) \\
                         & &  + P(A_1 \cap A_2 \cap A_3)
\end{eqnarray}

1. When we add (include) $P(A_1) + P(A_2) + P(A_3)$ we have added the pairwise intersection area once too many. So in the second line we subtract (exclude). But in doing so we subtracted $A \cap B \cap C$ out completely. So we add (include) $A \cap B \cap C$. Due to the aternating addition and subtraction, the process is called the inclusion-exclusion principle.  

2. Collectively, the terms  in  the second line can be written as $\sum_{i < j}P(A_i \cap A_j)$ for $i \in \{1, 2, 3\}$ and $j \in \{1, 2, 3\}$.  

3. Also, note that the cardinality of the set $\{A_i \cap A_j\} \text{such that} ~i < j$ is $\binom{3}{2}$. 

4. Note that $\{A_i \cap A_j\}$ and $\{A_i A_j\}$ **mean the same thing** and are used interchangeably. 



Let's consider 4 events $A_1, A_2, A_3, A_4$. 

\begin{eqnarray}
P(A_1 \cup A_2 \cup A_3 \cup A_4) &=&  P(A_1) + P(A_2) + P(A_3) + P(A_4) \\
						 & &  - P(A_1 \cap A_2) - P(A_1 \cap A_3) - P(A_1 \cap A_4) - P(A_2 \cap A_3) - P(A_2 \cap A_4) - P(A_3 \cap A_4)\\
                         & &  + P(A_1 \cap A_2 \cap A_3) + P(A_1 \cap A_2 \cap A_4) + P(A_1 \cap A_3 \cap A_4) + P(A_2 \cap A_3 \cap A_4) \\
                         & &  - P(A_1 \cap A_2 \cap A_3 \cap A_4)
\end{eqnarray}




Now we consider  the general case. If ${A_1}, {A_2}, \ldots, {A_n}$ are $n$ events, then

\begin{eqnarray}
  P(A_1 \cup A_2 \cup A_3 \cup \ldots \cup A_n) &=& \sum_{i=1}^{i=n} P(A_i)  \\
                                                & & - \sum_{{i_1} < {i_2}} P(A_{i_1} A_{i_2})\\
                                                & & +(-1)^{r+1}\sum_{{i_1} < {i_2} < ... < {i_r}} P(A_{i_1} A_{i_2}...A_{i_r})  + \ldots   ~~~~~~~~~ r = 3, \ldots, n-1 \\
                                                & & +(-1)^{n+1} P(A_1 A_2 A_3 \ldots A_n)
\end{eqnarray}

where ${i_1} < {i_2} < ... < {i_r}$ refers to all possible r-wise combinations of $(1, 2, \ldots , r)$. 


**What the above equation states is that the probability of the union of $n$ events is equal to the sum of the probabilities of the of events taken one at a time, minus the probability of the intersection of events taken two at a time, plus the probability of the events taken three at a time, and so on until the probability of the intersection of events all taken together.**


## Proof of the Inclusion-Exclusion Principle

The inclusion-exclusion identity can be proved by induction. Below we show one step of the induction. In partiucular,  assuming that the equation is true for $n = 2$, we derive the expression for $n = 3$. That is we derive the expression for $P(A_1 \cup A_2 \cup A_3)$  given that the basis step is true: $P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$. In doing this we will convince ourselves that  the above general Equation \@ref(eq:inclusion-exclusion-g) is true. 

We are given that for any two events $E_1$ and $E_2$, it is the case that $P(E_1 \cup E_2) = P(E_1) + P(E_2) - P(E_1 E_2)$.

Now consider three events $E_1$, $E_2$, and $E_3$. $E_1 \cup E_2 \cup E_3$ can be written as $E_{12} \cup E_3$ where $E_{12}$ is $E_1 \cup E_2$.
\begin{eqnarray}
P(E_1 \cup E_2 \cup E_3) &=& P(E_{12} \cup E_3) \\
                            &=& P(E_{12}) + P(E_3) - P(E_{12} E_3) \\
                            &=& P(E_1) + P(E_2) - P(E_1 E_2) + P(E_3) - P(E_{12} E_3)
\end{eqnarray}

Now, $E_{12} E_3 = (E_1 \cup E_2)\cap E_3$ which using the **distributive law** is $(E_1 \cap E_3) \cup (E_2 \cap E_3)$. Now we can use the inclusion-exclusion identity for two events to obtain the following

\begin{eqnarray}
P(E_{12} E_3) &=& P((E_1 \cap E_3) \cup (E_2 \cap E_3))\\
                &=& P(E_1 E_3) + P(E_2 E_3) - P(E_1 E_3 E_2 E_3) \\ 
                &=& P(E_1 E_3) + P(E_2 E_3) - P(E_1 E_3 E_2)
\end{eqnarray}

The last step is because $E_3 \cap E_3 = E_3$.  Now replace $P(E_{12} E_3)$ in Equation \@ref(eq:inclusion-exclusion-3i) to obtain the final result.


## Example: Network Recieve Buffer
A server has 10 incoming packet flows (flows are distinct and uniquely identified by a 5-tuple flow-id). Packets from each flow are stored in a buffer. Suppose there are two packets from each flow (packets have sequence numbers and hence are distinguishable) which are randomly assigned to locations in the buffer of size 20. What is the probability that no two packets from the same flow are in adjacent locations in the buffer?   

![Network-IO-Buffer](./Figures/network-io-buffer.jpg)

**Solution:** 

Let $E_i, i= 1, 2, 3, \ldots, 10$, be the event that packets from the ith flow are placed in adjacent locations in  the buffer. Then the required probability $\alpha$, that no two packets from the same flow are in adjacent locations, is given by 
\begin{equation*}
\alpha = 1 - P(\bigcup_{i=1}^{10} E_i)
\end{equation*}

Why? 

We use the first property of probability to obtain the expression above for $\alpha$. 

What we want is  $P(\bar{E_1} \bigcap \bar{E_2} \ldots \bigcap \bar{E_{10}})$. This, by **de Morgan's law**, is $P(\overline{{E_1} \bigcup {E_2}  \ldots \bigcup {E_{10}}})$. 

Now reflect on the fact that the events are not non-overlapping. Hence, using the Inclusion-Exclusion principle, we have 
\begin{equation*}
P(\bigcup_{i=1}^{10} E_i) = \sum_1^{10} P(E_i) + \ldots + (-1)^n \sum_{i_1 < i_2 < \ldots < i_n} P(E_{i_1} E_{i_2}  \ldots E_{i_n}) 
+ \ldots - P(E_1 E_2 \ldots E_{10})
\end{equation*}
Now we have to find the probabilities, i.e., $P(E_{i_1} E_{i_2}  \ldots  E_{i_n})$ which is the joint probability that packets from all of $n$ flows are placed adjacent to each other. Now the total number of ways in which 20 packets can be arranged in the buffer locations is $20!$ and these are equally likely outcomes. If we consider the packets from the flows that are adjacent as single location elements, then the total number of possible orderings is $(20 - n)!$. Thus, 

\begin{equation*}
P(E_{i_1} E_{i_2}  \ldots E_{i_n})  = \binom{10}{n} 2^n \frac{(20 -n)!}{20!}
\end{equation*}

The first term in the RHS is the number of ways we can choose the $n$ flows. The second term comes from the fact that there are two ways in arranging the two packets in each flow that are together.  Thus we get

\begin{eqnarray}
P(\bigcup_{i=1}^{10} E_i) &=&  \binom{10}{1} 2^1 \frac{19!}{20!} -  \binom{10}{2} 2^2 \frac{18!}{20!} +  \binom{10}{3} 2^3 \frac{17!}{20!} - \ldots - \binom{10}{10} 2^{10} \frac{10!}{20!} \\[2mm]
                          &\approx& 0.6416608
\end{eqnarray}

Let's write the code to compute it. 

```r
sum = 0  
n = 10
for (i in 1:n)
{
  sum = sum + (-1)^(i+1)*(factorial(n)/(factorial(n - i)*factorial(i)))*(2^i)*factorial(2*n - i)/factorial(2*n)
}
print(sum)
```

Thus, $\alpha = 1 - 0.6416608 =  0.3583392$

# III.B. Matching Problem

There is a deck of $n$ cards labelled $1, \ldots, n$. The  cards are shuffled. A person calls out numbers 1 through $n$ in sequence and with each call flips a card from the deck. What is the probability that for at least one called number, the corresponding card number will match? This problem was first considered by Pierre-Remond  Montemort's  around 1713.  

Another version of the problem is called the **Hat matching problem**.
Suppose $n$ people throw their hats in a room. Then they randomly pick a hat. What is the probability that at least one person picks his own hat.  Here we use the language of cards.


Let $A_j$  be the event that the $j$th card matches.

\begin{eqnarray}
P(\text{at least 1 match}) &=& P(A_1 \cup A_2 \cup A_3 \cup \ldots \cup  A_n) \\
              &=& \sum_{i=1}^{i=n} P(A_i)  \\
              & & - \sum_{{i_1} < {i_2}} P(A_{i_1} \cap A_{i_2})\\
              & & +  (-1)^{r+1}\sum_{{i_1} < {i_2} < ... < {i_r}} P(A_{i_1} \cap A_{i_2} \cap \ldots \cap A_{i_r}) + \ldots \\
              & & + (-1)^{n+1} P(A_1 \cap A_2 \cap A_3 \cap \ldots \cap A_n)
\end{eqnarray}

The above is based on the Inclusion-Exclusion principle.  We will now determine each of those terms. 

1. The single terms 
\begin{equation}
P(A_j) = \frac{1}{n}  ~~~~~ \text{for all $j = 1, \ldots, n$}
\end{equation}
This comes from the observation that all positions are equally likely for card labelled $j$.  
How many such terms are there? Answer: $n$ which we could also write as $n \choose 1$. 


2. Terms taken two events at a time
\begin{equation}
P(A_i \cap A_j) = \frac{(n-2)!}{n!}  ~~~~~ \text{for all} ~~~ i, j, = 1, \ldots, n ~~~~\text{and}~~~~ i > j
\end{equation}
This is based on the observation that 2 cards $A_i$ and $A_j$ are in the right positions  so the remaining $n - 2$ can be permuted in $(n - 2)!$ ways which gives the total number of favorable outcomes, and note that $n!$ is the total number of outcomes.  
How many such terms are there? Answer: there are  $n \choose 2$ terms. 


3. We can generalize this to $k$ events taken together 
\begin{equation}
P(A_{i1} \cap A_{i2} \ldots \cap A_{ik}) = \frac{(n-k)!}{n!}  ~~~~~ \text{for all}~~~ i_1, i_2, \ldots i_k = 1, \ldots, n ~~~~\text{and}~~~i_1 > i_2 > \ldots > i_k
\end{equation}
How many such terms are there? Answer: There are  $n \choose k$ terms. 


4. 
Now we put it all together

\begin{eqnarray*}
P(\text{at least 1 match}) &=& \binom{n}{1} \times \frac{1}{n}  \\
                           & & - \binom{n}{2} \times \frac{(n-2)!}{n!}  \\
                           & & +(-1)^{k+1} \binom{n}{k} \times \frac{(n-k)!}{n!}   ~~~~ k = 3, \ldots, n-1 \\
                           & & +(-1)^{n+1} \binom{n}{n} \times \frac{(n-n)!}{n!} \\
                           &= & \frac{1}{1!} - \frac{1}{2!} + \frac{1}{3!} + \ldots + (-1)^{n+1} \frac{1}{n!} \\
                          &\approx& 1 - \frac{1}{e}
\end{eqnarray*}

A couple of notes:

1. To get the final result we use the asymptotic value of the Taylor series as $n \rightarrow \infty$. Recall that $$e^x = 1 + \frac{x}{1!} + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots$$ 
2. Why is there an asymptotic value? As  $n \rightarrow \infty$, the chance of getting a match becomes smaller and smaller however, the number of attempts for a match becomes larger and larger. These two opposite forces converge to $1 - \frac{1}{e}$.

In order to save power we need to put the relay nodes to sleep. In which case how can they synchronize? 



The problem relates to synchronizing 2 sensor nodes while minimizing sensor battery power usage. We consider that time is divided into fixed length frames with each frame containing $n$  fixed length slots. In each frame, each sensor node independently and randomly selects $m$ slots to be awake (and sleeps the remaining $n-m$ slots). Let $P(\text{match})$ denote the probability that in a frame there is at least one common slot in which both the sensors are awake. We want to find $P(\text{match})$


![Two-Sensor-Synchronization](Figures/Sensor-synchronization.jpg)


1. If $m > n/2$ then the $P(\text{match}) = 1$. This makes sense: if each node selects greater than half of the number of slots then, there will be at least one match. (This is an example of the Pigeonhole principle.)


2. If $m \leq n/2$ then the probability that two sensor nodes will not have a single slot in common is given by

\begin{equation*}
  P(\text{no match}) = \frac{{n \choose m} {n-m \choose m}}{{n \choose m}^2}
\end{equation*}

Here the demoninator is the size of the sample space, $\left| S \right|$, since the sensors are independent and each one is active in $m$ of the $n$ slots. 

The numerator is the number of ways the sensors can choose non-overlapping slots.  The first has $n$ slots to choose from, the second has $n-m$ slots to choose from. 

Thus,

\begin{equation*}
     P(\text{at least one match}) = 1 - P(\text{no match})
\end{equation*}

```r
lp0 = 80 
active = 160
rx = 1500
tx = 3000
n = 100      # The number of slots in a timeframe
m = 40       # The maximum number of active slots for each sensor 

power = numeric(m)
match = numeric(m)

for (i in 1:m){
    match[i] = 1 - choose(n,i)*choose(n-i,i)/(choose(100,i))^2
    power[i] = i*(3000+1500+160) + (n-i)*80
}
plot(power, ylab="Power", xlab="Number of slots awake")
plot(match, ylab="Probability of match", xlab="Number of slots awake")
value = 0.632        # this is the asymptotic solution to the matching problem, 1-(1/e)
abline(h=value)
abline(v=20)
```

# IV. Conditional Probability

We started with the nave definition of probability and solved some challenging and important problems. We discussed the limitations of the nave definition and then gave a formal definiton of probability. Now we take the next step to increase the power of probabilistic modeling. In particular, we discuss how we update  the evaluation of the uncertainty (risk) when we are given new data. In a sense this is what we as humans are constantly doing--re-evaluating risk based on data that we sense from the environment.  Conditional probability is also considered the foundation for statistics. 

***Definition of Conditional Probability***

\begin{equation*}
    P(A|B) = \frac{P(A \cap B)}{P(B)}  ~~~~~~~~~~~ P(B) > 0
\end{equation*}

In Figure \@ref(fig:cond) given that event $B$ has occurred the relevant portion of event $A$ is $A \cap B$. Given that $B$ has occurred, outcomes that are not in $B$ ($B^{\mathsf(c)}$)are no longer relevant. The sample space has now collapsed to $B$ since only outcomes in $B$ are possible. 


![Conditional Probability](Figures/Conditional-probability.jpg)



### Few Theorems

1. **Theorem 1**
\begin{eqnarray}
  P(A \cap B) &=& P(B) \times P(A|B) \\
  			  &=& P(A) \times P(B|A)
\end{eqnarray}

Note that $P(A \cap B) = P(B \cap A)$,  but also note that $P(A|B) \ne P(B|A)$ in general. 

2. **Multiplication Rule**
\begin{equation*}
    P(A\cap B \cap C) = P(A) \times P(B|A) \times P(C|A \cap B)
\end{equation*}


3. **Bayes' Rule**  
\begin{equation*}
   P(A|B) =  \frac{P(A) \times P(B|A)}{P(B)}
\end{equation*}  
$P(A|B)$ is called the **posterior probability** of A and $P(A)$ is called the **prior**. Note that $P(B)$ can also be written as  

\begin{equation}
  P(B) = P(A) \times P(B|A) + P(A^\mathsf{c}) \times P(B|A^\mathsf{c})
\end{equation}


4. **The Law of Total Probability**: If $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events, then

\begin{equation*}
        P(B) = P(B|A_1)\times P(A_1) + P(B|A_2)\times P(A_2) + \ldots + P(B|A_n)\times P(A_n)
\end{equation*}


The above theorems can be  proven using the axioms of probability and the definition of conditional probability. 



## Examples

***Example 1: Rolling Two Dice***  
Two fair six-sided dice are rolled. What is the conditional probability that one of them lands on a 6, given that the dice land on different numbers?

![Sample-space-with-two-dice](Figures/Sample-space-with-two-dice.jpg)


The restricted sample space consists of 30 outcomes. $S = \{(i, j), i \neq j, i, j \in \{1,2, 3, 4,5, 6\}\}$. This can be seen by considering how many outcomes have both dies equal -- there are only 6. There are then ten outcomes in which at least one lands on 6. Thus, the required probability is 10/30 = 1/3.

**NOTE:** All the laws of probabilty still hold, but the sample space is now confined to a smaller subset of the original!

***Example: Testing for a Rare Disease***  
Patients get tested for a disease that affects 1\% of the population. The test is 95\% accurate. If a patient tests positive what is the probability that the patient has the disease?  
A common (wrong) answer is 95\%.  
Let  
1. ***T***: event that a patient tests positive (which is conditional, as shown below).   
2. ***D***: event  that a patient has the disease, so $P(D) = 0.01$  

We will assume that the test being 95\% accurate implies that

\begin{eqnarray*}
P(T|D) &=& 0.95 \\
P(T^\mathsf{c} |D) &=& 0.05 \\
P(T^\mathsf{c} |D^\mathsf{c}) &=& 0.95 \\
P(T |D^\mathsf{c}) &=& 0.05
\end{eqnarray*}

Let's make sure we understand the above terms. The terms are a formal re-statement of the properties of the test. Recall that "95\%" accurate implies that the test is 95\% accurate in both directions--for testing the presence or absense of the disease. Hence, in this scenario, 

1. $P(T|D)$ can be thought of as the **"true positive"** conditional probability. For example, suppose that the person tested positive for the disease and then later presented symptoms in a clinical setting. 


2. $P(T^\mathsf{c} |D)$ is the **"false negative"** conditional probability. Suppose in this case that the person tested negative for the disease, but then presented symptoms later in a clinical setting. 


3. $P(T^\mathsf{c} |D^\mathsf{c})$ is the **"true negative"** conditional probability. This is the odds that someone tested negative and does not, in fact, have the disease. 


4. $P(T |D^\mathsf{c})$ is the **"false positive"** conditional probability. This patient tested positive for the disease, but never presented symptoms so they do not have the disease.  


We now use Bayes' Theorem to determine the probability that the patient has the disease given that they tested positive: 

\begin{eqnarray*}
P(D|T) &=& \frac{P(D) \times P(T|D)}{P(T)} \\[2mm]
       &=& \frac{P(D) \times P(T|D)}{P(D) \times P(T|D) + P(D^\mathsf{c}) \times P(T|D^\mathsf{c}) } \\[2mm]
        &=& \frac{0.01 \times 0.95}{0.95 \times 0.01 + 0.05 \times 0.99 } \\[2mm]
       &=& 0.16
\end{eqnarray*}

Note: The numerator uses Theorem 1. The denominator uses the law of total probability. 

Here is another way of arriving at the solution: Let say we have a population of 10,000. Since 1\% of the population has the disease, this implies that about 100  people will have the disease and 9900 people will not have the disease. Even though the test is reasonably accurate, 5\% of 9900 (which is 495) people who do not have the disease will test positive. And 95 of the hundred people who have the disease will test positive. Now if a person tests positive, the probability that she has the disease is $\frac{95}{95+495}$ which is 16\%. This shown in the following figure 

![Test-for-disease](Figures/Test-for-disease.jpg)


The important point to note here is that even though the test is reasonably accurate, a significant number who do not have the disease will test positive. You can also think of this as a "signal to noise" problem. The "signal" is the fact that this only affects 1\% of the population, but the noise is 5\% of inaccuracy of the test. 

This contradicts **the common fallacy** that $P(A|B) = P(B|A)$ with the introduction of new information. Notice how the fact that the disease only affects 1\% of the population was subtly conveyed, but indeed was key to calculating $P(D)$.

### IV.A. The Law of Total Probability

In this case we use conditional probability to determine the probability $P(B)$ of an event B. If $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events, then

\begin{eqnarray*}
        P(B) &=& P(B \cap A_1) \cup P(B \cap A_2) \cup \ldots \cup P(B \cap A_n) \\
            &=& P(B \cap A_1) + P(B \cap A_2) + \ldots + P(B \cap A_n) \\
            &=& P(B|A_1)\times P(A_1) + P(B|A_2)\times P(A_2) + \ldots + P(B|A_n)\times P(A_n)
\end{eqnarray*}

This is best illustrated in Figure \@ref(fig:Law-of-total-probability).


![Law-of-total-probability](Figures/Lawoftotalprobability.jpg)



It is important to note  that $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events.  Note that a simple case is if we consider $A$ and $A^\complement$.

\begin{eqnarray*}
        P(B) &=& P(B \cap A) \cup P(B \cap A^\complement)\\
            &=& P(B \cap A) + P(B \cap A^\complement) \\
            &=& P(B|A)\times P(A) + P(B|A^\complement)\times P(A^\complement)
\end{eqnarray*}

The last line follows from Theorem 1, that $P(C \cap D) = P(D) \times P(C|D)$.

### Example: A Conditional Urn Problem

This following problem is difficult and shows the use of the Law of Total Probability  and Bayes' Theorem. As you go through it, remember that $P(\ast|F)$ is a complete probability space and that all axioms of probability apply. Only the sample space is now changed, it is reduced to the set that remains after that event $F$ has happened.


Here is the problem statement:  There is a one ball in a bag. It is equally likely that the ball is either red or blue. A red ball is now added to the bag and then a random ball is taken out. If the ball withdrawn is red what is the probability that the remaining ball is also red?

We define the following events

1. A: event that the initial ball is red.
2. B: event that  the removed ball is red.
3. C: event that the remaining ball is red.  (Which logicaly means the initial ball was red if event B happens.)


We need to find $P(C|B)$. To find this by conditioning whether the initial ball is red.  (See a more explicit derivation below the figure.) 

\begin{eqnarray*}
  P(C|B) &=& P(C|B \cap A) P(A|B) + P(C|B \cap A^\complement) P(A^\complement |B) \\
         &=& 1 \times  P(A|B) + 0 \times   P(A^\complement|B)\\
         &=& P(A|B)
\end{eqnarray*}


To find $P(A|B)$ we use Bayes' Rule

\begin{eqnarray*}
  P(A|B) &=& \frac{P(A) P(B|A)}{P(B)} \\[1mm]
         &=& \frac{(\frac{1}{2}) \times 1}{P(B|A) P(A) + P(B|A^\complement) P(A^\complement)} \\[1mm]
         &=& \frac{\frac{1}{2}}{\left(1 \times \frac{1}{2}\right)  + \left(\frac{1}{2}\times \frac{1}{2}\right)} \\[1mm]
         &=& \frac{2}{3}
\end{eqnarray*}


Thus the required probability $P(C|B) = 2/3$.

Note the denominator in the final equation follows since $A$ means event $B$ samples from $\{R,R\}$, and $A^C$ means $B$ samples from $\{B,R\}$. 

Here is a figure of the original sample space and the reduced sample space conditioned on event B, and the reduced sample space conditioned on event C: 

<img src="./Figures/conditional_urn.jpg" width="800">

From the figure we can see that $P(C|B) = \frac{|C\cap B|}{|B|} = 2/3$ since the size of the reduced sample space conditioned on event B is of size 3, and the event C covers 2 of those three outcomes. 

***Explicit derivation*** of the formula above

\begin{eqnarray*}
P(C|B) &=& P[C|(B\cap A + B\cap A^c)] \\[2mm] 
       &=& \frac{P[C\cap(B\cap A + B \cap A^c)]}{P(B)}\\[2mm] 
       &=& \frac{P[(C\cap B \cap A) + (C\cap B \cap A^c)]}{P(B)} \\[2mm]  
       &=& \frac{P(C\cap B \cap A)}{P(B)} \\[2mm]  
       &=& \frac{P(B) P(A|B) P(C| A\cap B)}{P(B)}\\[2mm]  
       &=& \frac{P(B) P(A|B)}{P(B)}\\[2mm]  
       &=& P(A|B)
\end{eqnarray*}

- The first line follows since $B=B\cap A + B\cap A^c$. 

- The second line follows since $P(E|F) = P(E\cap F) / P(F)$ and since $P(B\cap A + B\cap A^c) = P(B)$ from the sentence above. 

- The third line follows from the distributive law for sets. 

- The forth line follows since $(C\cap B \cap A^c)=0$.  In other words if events $A^c$ and $B$ both happen then we know that $C$ cannot happen and instead $C^c$ must have happened (i.e., the remaining ball is blue). 

- The fifth line follows from the multiplication law. 

- The sixth line follows since $P(C | B \cap A)=1$.  In other words if events $A$ and $B$ both happen then we know event $C$ must happen (i.e., the remaining ball is red).  


### Example: Prosecutor's Fallacy
Police in a city of 1 million ($10^{6}$) are working on a crime. Police are given some description of the perpetrator (such as height, limp, etc) and based on statistics 1 in 10,000 people satisfy the description. Police on a routine patrol see a person fitting the description  and they make the arrest. The prosecutor makes the following case:

1.  S1: Only 1 in 10,000 fit the description;
2.  S2: It is highly unlikely that an innocent person fits the description;
3.  S3: It is highly unlikely that the defendant who fits the description is innocent.

We want to clearly understand where the prosecutor is making a wrong statement.

Statement S1, which states the statistic that 1 in 10,000 people fit the description, is given to be true. 
With $10^{6}$ people, there will be 100 people who will fit the description only one of whom is guilty. So the probability that the defendant is guilty is 1 in 100 which is small. 

Statement S2 is calculating $P(D|I)$ is the probability that an innocent person fits the description.
Let $D$ denote the event that a person fits the description and let $I$ denote the event that a person is innocent. The set $D$ will contain 100 people (1 of whom is guilty). The set $I$ will contain 999,999 people (since only one is guilty). Thus the set $D\cap I$ has 99 people. 


\begin{eqnarray*}
P(D|I) &=& \frac{P(D \cap I)}{P(I)} \\[1mm]
       &=& \frac{99}{999,999} \\[1mm]
       &=& 9.9 \times 10^{-5}
\end{eqnarray*}

This is indeed small and hence statement S2 is true.

Statement S3 is asking the probability that the defendant is innocent given that the defendant fits the description which is $P(I|D)$. The prosecutor is claiming that this is really small and hence the defendant is guilty. We can calculate it

\begin{eqnarray*}
P(I|D) &=& \frac{P(D \cap I)}{P(D)} \\[1mm]
       &=& \frac{99}{100} \\[1mm]
       &=& 0.99
\end{eqnarray*}

which is very close to 1.

Hence, statement S3 is wrong. The prosecutor in essence is wrongly equating $P(I|D)$ to be the same as $P(D|I)$ but they are not.

Read about the real case of [Sally Clark](https://en.wikipedia.org/wiki/Sally_Clark), who was convicted in 1999.

## IV.B. Independence of Events 

We  consider two events $A$ and $B$ to be independent if 

\begin{equation}
P(A \cap B) = P(A) \times P(B) 
\end{equation} 


What this intuitively means is that that the occurence of Event A does not affect the occurrence of Event B. It is important to note that this is different from Events A and B being mutually exclusive. In fact, if two events are mutually exclusive they are not independent. 


***Example: Two Randomly-Drawn Cards***  
Suppose a card is drawn at random from a deck of cards. Let A be the event that the card is a spade and B be the event that card is an Ace. These two events are independent. $P(A) = 13/52 = 1/4$ and $P(B) = 4/52 = 1/13$ and $P(A \cap B) = 1/52 = P(A) \times P(B)$


***Example: Odds of Having a Baby Boy or Girl***
Consider a family of two children. Let $A$ be the event that the family has children of both genders and $B$ be the event that there is at most one girl. Are the events $A$ and $B$ independent? 

The sample space $S = \{ bb, bg, gb, gg\}$. Now $P(A) = 2/4 = 1/2$ and $P(B) = 3/4$. $P(A \cap B) = 2/4 = 1/2$ which is not equal to $P(A) \times P(B)$. So no, the events are not independent. 


## Practice Problems 


3 cards are drawn from a deck of 52 cards in succession without replacement.  We define the following events


1. $A_1$: the first card is a Red Ace
2. $A_2$: the second card is a 10 or a Jack
3. $A_3$: the third card is a greater than 3 but less than 7.


Find $P(A_1 \bigcap A_2 \bigcap A_3)$.


In general if in an experiment events $A_1, A_2, \ldots , A_k$ can occur then
\begin{equation*}
  P(A_1 \cap A_2 \cap \ldots \cap A_k) = P(A_1) ~ P(A_2 | A_1) ~ P(A_3|{A_2 \cap A_1}) ~ \ldots P(A_k | A_{k-1} \cap A_{k-2} \ldots \cap A_1).
\end{equation*}

This is called the multiplication rule. You can prove this using induction.
We can use the multiplication rule to solve our problem. First,  $P(A_1) ~=~ 2/52$, second, $P(A_2|A_1) ~=~ 8/51$, and finally, $P(A_3|A_1 \cap A_2) ~=~ 12/50$. Then,

\begin{eqnarray*}
  P(A_1 \cap A_2 \cap A_3) &=& P(A_1) ~ P(A_2 | A_1) ~ P(A_3|{A_2 \cap A_1}) \\
    &=& (2/52) \times (8/51) \times (12/50) \\
    &=& 8/5525
\end{eqnarray*}
