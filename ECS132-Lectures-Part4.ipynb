# VII. Continous Random Variables 

The last distribution for a discrete random variable that we covered was the Poisson distribution. There the random variable $X \in \mathbb{Z}$, but the parameter $\lambda \in \mathbb{R}$. The parameter $\lambda$ = (rate)$\cdot$(time), and both rate and time can be real numbers. Now, we enter fully into the set of real numbers, considering random variables which can take on any real number.

Like discrete random variables, continuous random variables are functions that map the outcomes of a sample space to the $\mathbb{R}$. This is shown in Figure \@ref(fig:continuous-random-variable) below. The difference is that the sample space now is ***uncountably infinite*** and they map to real values.


![continuous-random-variable](Figures/Continuous-random-variable.jpg)


Let $X$ denote the random variable and takes values $x \in \mathbb{R}$. Remember $X$ is the **name** of the random variable and $x$ is a real number describing the outcome of a random trial. The sample space 

$$ S = \{ -\infty < x < \infty\}$$

Unlike the discrete case, we now have a problem in quantifying $P(X = x)$. In particular, $P(X = x)$ must be equal to  0. Formally stated:

\begin{equation}
P(X=a) = \int_a^a f(x)dx = 0
\end{equation}

This is because, even in an arbitrary small neighborhood of $x$ there are infinite number of values so the probability that $X$ $\text{exactly}$ takes the value $x$ will be 0.  Looking at it in another way if we assign a positive value  to $P(X = x)$, however small, since $x$ can take uncountably infinite number of values, then the sum of the probabilities will become infinite. Instead we consider a region $dx$ around $x$, so $p(x)dx = P\left[x \le X \le (x+dx)\right]$ discussed in more depth below. 

In summary, although the probability that the continuous random variable takes a specific value is 0 we can quite easily define the probability that it is within a *range of values.* To make things concrete consider the temperature of Davis at 12 noon. Note that we are assuming that we have an measuring instrument that can measure the temperature with arbitrary precision.

![temperature-profile](./Figures/Temperature-profile.jpg)

In order to work with continuous random variables, what will  make sense is to define probability for a range of values. In order to do that, we define a function $f_X (x)$ which we call the **probability density function (pdf)**. 
Note it is intentional notation that the pdf is denoted $f_X(x)$ since the **cumulative distribution function (cdf)**, in analogy to discrete variables, is denoted $F_X(x)$. 

We first give  the following definition:  $f_X (x)$ is a function of $x$ that when multiplied by a small interval $\Delta x$ gives the probability that $X$ lies between $x$ and $x + \Delta x$.

\begin{equation}
	P(\text{Temperature lies in the range $x$ and $x + \Delta x$}) \approx f_X(x) \Delta x
\end{equation}

Note that the left hand side is a probability. $f_X (x)$ is not a probability; we get the probability when it is multiplied by a range $\Delta x$. Note that the right hand side gives an area (see the shading in the figure above).  Some observation:

1.  For a given value of temperature say $t = 95$ and $\Delta = 2$, the probability is  $0.005*2 = 0.01$. If we increase the $\Delta = 4$, then to probability is higher $0.005*4 = 0.02$. This makes sense since it is more likely for the temperature to fall in a larger range.


2. For the same range the probability is different for different values of the temperature. The probability that temperature is between 60 and 62 is $0.001 * 2 = 0.002$. Similarly, the probability that the temperature is between 82 and 84 is 0.014. The density is different at  different temperature values. The likelihood of the temperature being around 60 is low hence the probability density around 60 is low. The probability of the temperature being around 85 is high so the density around 85 is high.  The probability over a fixed range is higher where the density is higher.   

The above definition is an approximation and is acceptable as long as the the change in the function over $\Delta x$ is small compared to the value of $f_X(x)$.

In order to extend the definition of probability to make it more rigorous, we consider the area under the curve instead of the rectangles. In other words we consider the **integral** of $f_X (x)$ between  $x$ and $x + \Delta x$. A more rigorous definition of pdf is as follows: $f_X (x)$ is a function of $x$ for which the area under the curve $f_X (x)$ between  $x$ and $x + \Delta x$ gives the probability that $X$ lies between $x$ and $x + \Delta x$.  In the above definition $\Delta x$ need not be small.


## VII.1. Probability Density Function (PDF) 

If $X$ is a random variable, $f_X(x)$ will denote its density function, which has the following properties:

1. $f_X(x) \geq 0$. This clearly follows from the Axiom that probability is a positive quantity between 0 and 1.


2. $\int_{-\infty}^{\infty} f_X(x) dx  = 1$. This is the normalization condition and is the probability of the entire sample space $S$.  

***Example:***

Suppose we have some Random Variable $X$, whose density function is given by:

\begin{equation}
f_X(x) = 
\begin{cases}
cx~~ \text{for} ~~0 \leq x \leq 2 \\
0~~otherwise
\end{cases}
\end{equation}
For what value of $c$ is $f_X(x)$ a valid density function?

**Solution:** 

We require that $f_X(x) \geq 0$ for $0 \leq x \leq 2 $. Hence, $c > 0$. Furthermore, from the properties of density functions, we know that:

\begin{equation}
\int_{-\infty}^{\infty}f_X(x)dx = 1
\end{equation}

So we will take the integral and set $c$ accordingly:

\begin{equation}
\int_{0}^2 cxdx = \frac{cx^2}{2} \bigg\rvert_0^2 = \frac{c4}{2} = 2c
\end{equation}

So it will be density function if $c=\frac{1}{2}$


## VII.2. Cummulative Distribution Function (CDF) 
The definition of the CDF is the same as in the case  of the discrete random variable

\begin{eqnarray*}
F_X (y) &=& P(X \leq y) \\
        &=& \int_{-\infty}^y f_X(x) dx
\end{eqnarray*}

Recall that $x$ and $y$ are just variable names, and since it is customary to think of integrating over variable $dx$ we use $y$ to denote the cutoff value for $F_X$.  And similar to the case of the discrete random variable

\begin{eqnarray*}
P(a < X \leq b) &=& F_X (b) - F_X(a)  \\
                &=& \int_{a}^{b} f_X (x) dx
\end{eqnarray*}

In continuous random variables $<$ (or $>$) and $\leq$ ($\geq$) are the same since the probability  that the random variable takes a specific values is equal to 0.

## VII.3. Mean and Variance

Recall that for a discrete rando variable $E(X) = \sum_j x_j p_j$. In analogy, the expected value of a continuous random variable  $E(X)$ is given by

\begin{equation}
E(X) = \int_{-\infty}^{\infty} x f_X(x)dx
\end{equation}

and the Variance

\begin{eqnarray*}
\text{Var}(X) & =& \int_{-\infty}^{\infty} (x - E(X))^2 f_X(x) dx\\[2mm]
&=& E(X^2) - \left(E(X)\right)^2
\end{eqnarray*}

***Example:***  
$X$ is some continuous random variable with density function:

\begin{equation}
f_X(x) = \frac{1}{2}x~~~~~~ {\rm for}\ 0 \leq x \leq 2 \ \ {\rm and \ zero \ otherwise}
\end{equation}

Find $E(X)$, $E(X^2)$, and $Var(X)$.

**Solution:**  
\begin{align}
E(X) &= \int_0^2 x\frac{1}{2}xdx = \frac{x^3}{6} \bigg\rvert_0^2 = \frac{8}{6} = \frac{4}{3} \\[5mm]
E(X^2) &= \int_0^2 x^2\frac{1}{2}xdx = \frac{x^4}{8} \bigg\rvert_0^2 = 2 \\[5mm]
Var(X) &= E(X^2) - E(X)^2 = 2 - \frac{16}{9} = \frac{2}{9}
\end{align}


Finally, we can also define the moments as we did for the discrete random variable with the sum replaced by the integration.

## VII.4.  Uniform Distribution

A random variable $X$ follows a Uniform Distribution $$X \sim \text{Unif}(a,b)$$ then the **pdf** is given by

\begin{equation}
	f_X (x) =
	   \begin{cases}
	   	  \frac{1}{b - a} & \text{if $ a \leq x \leq b $} \\[2mm]
	   	  0               & \text{otherwise}
	   \end{cases}
\end{equation}


![uniform-distribution](Figures/uniform-distribution.jpg)

If $X \sim \text{Unif}(a,b)$, one way to think about this distribution is as follows. The probability that a randomly selected point within $(a, b)$  is in some interval is proportional to the length of the interval.



### VII.4.a) Properties
The following are some important facts:

1. $f_X (x) \geq 0$ for all $x$


2. $\int_{- \infty}^{\infty} f_X (x) dx = 1$  


3. The CDF $F_X (y)$ is given by
\begin{eqnarray*}
	   	F_X (y) &=& P(X \leq y) = \int_{- \infty}^{y} f_X (x) dx \\[2mm]
                &=& \int_{a}^y \frac{1}{b-a} dx = \frac{1}{b-a} \int_{a}^y dx \\[2mm]
	   	        &=& \frac{y - a}{b -a}\\[3mm]
\end{eqnarray*}

4. It is easy to show that  
\begin{equation}
E(X) = \frac{{b + a}}{2}
\end{equation}

5. It is easy to show that
\begin{equation}
\text{Var}(X) = \frac{(b - a)^2}{12}
\end{equation}




#### Example: Batch Job Run Times
The execution time of a certain class of jobs in the Edison supercomputer is a random variable $X$ which is Uniformly distributed over (1, 10), i.e., $X ~\sim \text{Unif}(1, 10)$.

1. What is the probability that a job of the given class will have a execution time of less than 3 mins?

**Solution:**  
\begin{eqnarray*}
		P(X \leq 3) &=& \int_{- \infty}^{3} f_X (x) dx \\[2mm]
		            &=& \int_{1}^{3} \frac{1}{10 - 1} dx \\[2mm]
		            &=& \frac{2}{9}
\end{eqnarray*}

2. What is the probability that a job of the given class will take more than 6 mins?

**Solution:**  
\begin{eqnarray*}
		P(X \geq 6) &=& \int_{6}^{\infty} f_X (x) dx \\[2mm]
		            &=& \int_{6}^{10} \frac{1}{10 - 1} dx \\[2mm]
		            &=& \frac{4}{9}
\end{eqnarray*}

## VII.5. Exponential Distribution

You are probably familiar with the exponential function $e^x$ which occurs pervasively in nature and we have used the Taylor Series expansion of $e^x$ several times in this course. 

Here we consider the exponential distribution. A random variable $X$ is exponentially distributed 

$$X \sim \text{Expo}(\lambda)$$ 

with $\lambda > 0$ where $\lambda$ is called the rate parameter. (Specifically $\lambda$ is given in units of: #-of-events/unit-time.)  The **pdf** is given by

\begin{equation*}
      f_X (x) = \left \{ \begin{array} {ll}
      \lambda e^{- \lambda x} &  x \geq 0 \\
      0 & {\rm otherwise}
      \end{array} \right.
\end{equation*}

The exponential distribution is typically **concerned with the amount of time until some specific event occurs**. For example, the amount of time (beginning now) until an earthquake occurs has an exponential distribution. Other examples include the length, in minutes, of long distance business telephone calls, and the amount of time, in months, a car battery lasts.  So for $X \sim \text{Expo}(\lambda)$ the random variable $X$ corresponds to the amount of time one has to wait for the occurrence of an event (e.g., an earthquake to happen, a phone call to end, the battery to die). 

#### Warm-up example: Check that this is a valid PDF
**Solution:**  

\begin{equation}
\int_{-\infty}^\infty f_X(x) dx = \int_0^\infty \lambda e ^{-\lambda x}dx = -e^{-\lambda x} \bigg\rvert_0^\infty = 0-(-1) = 1
\end{equation}

Remember $e^{0}=1$. This result above also puts us very close to finding the CDF of the exponential distribution. 

The **CDF** is:

\begin{equation}
F_X(y) = \int_0^y \lambda e ^{-\lambda x}dx = -e^{-\lambda x} \bigg\rvert_0^y = 1-e^{-\lambda y}
\end{equation}

###  VII.5.a) Properties

If $X \sim \text{Expo}(\lambda)$ show that

\begin{eqnarray*}
      E(X)  &=& \frac{1}{\lambda} \\[5mm]
      Var(X) &=& \frac{1}{\lambda^2}
\end{eqnarray*}
Hint: This requires a relatively simple integration by parts.

#### Example 1: Waiting time in a line

Suppose that a barrista at Starbucks spends on average 3 mins with each customer in line. What is the probabilty the barrista will take less than 6 mins with one customer? 

**Solution:**
Here $X$ is the random variable corresponding to the service time for each customer with rate parameter $\lambda$ = (1 customer/3 mins). 

$$P(X \le 6) = \int_0^6 \lambda e ^{-\lambda x}dx = 1-e^{-\lambda 6} = 1-e^{-2} = 0.865$$

Thus we also immediately know that $P(X > 6) = 0.135$.


### VII.5.b) Memory-less property of Exponential distribution

Consider a sequence of independent events (like earthquakes or phone calls) each with $X \sim \text{Expo}(\lambda)$. Suppose, we have observed that time $s$ has passed and the next event has not yet occurred. We are interested in finding the remaining  time until the next event. Let $Y$ denote this random variable. We would like to find $F_Y (t)$ which by definition is equal to $P(Y \leq  t)$.  We will find the complement $P(Y > t)$

\begin{eqnarray*}
    P(Y > t) &=& P(X > (t+s) | X > s)  \\[3mm]
                  &=&  \frac{P(X > (t+s)  \cap  X > s)}{P(X > s) }~~~~\text{Theorem 1 of conditional probability} P(A\cap B) = P(B|A)P(A)\\[3mm]
                  &=&  \frac{P(X > (t+s) )}{P(X > s) }~~~~\text{Intersection is simply t+s} \\[3mm]
                  &=& \frac{1 - (1 - e^{-\lambda (s+t)})}{1 - (1 - e^{- \lambda s})} \\[3mm]
                  &=& \frac{e^{-\lambda (s+t)}}{e^{- \lambda s}} = \frac{e^{-\lambda s}e^{-\lambda t}}{e^{- \lambda s}}\\[3mm]
                  &=& e^{-\lambda t}~~~~\text{Still exponential}
\end{eqnarray*}

And $P(Y \le t) = 1 - e^{-\lambda t}$, the CDF for the exponential distribution. This shows that the distribution of $Y$ is the same as $X$. Both are $\text{Expo}(\lambda)$. The Exponential distribution is the only continous distribution that has the memory-less property.  

**Bottom line:** This means that the expected waiting time is ${\rm Expo}(\lambda)$ irrespective of the moment that we first started counting. 


#### Relation to Poission processes

This is very much related to the Poisson process. Recall for the Poisson process the random variable $X$ is the number of events that happen in a specified time window. We recap here a calculation from that section. Consider events that follow a Poisson process with rate $r$ events per unit time (so $\lambda^{\rm Poiss} = r\cdot t$, the unit rate times the length of the time window). Let $Y$ be a random variable that denotes the waiting time between events in the same time window; i.e., starting from an event what is the time until the next event. This is also called the inter-arrival time.

![poisson-process](Figures/Poisson-process-2-update.jpg)


\begin{eqnarray*}
P(Y \leq t) &=& 1 - P(Y > t) \\
                  &=& 1 - \text{Probability there no arrivals in time $t$} \\
                  &=& 1 - P(Y = 0)  \\
                  &=& 1 - e^{- r t}
\end{eqnarray*}

the CDF for the exponential distribution. So $X\sim {\rm Expo}(r)$. 

Note the Poisson process considers the number of events in a specified time window so the parameter $\lambda^{\rm Poiss} =$ (rate) $\cdot$ (time-window). In the exponential distribution instead the paramater $\lambda^{\rm exp}$ is the rate per unit time (i.e., the number of events observed divided by the length of the time window). In summary:

- For Poisson, $\lambda$ is the number of events in a specified time window.

- For exponential, $\lambda$ is the number of events per unit time. (i.e., set the time window of consideration to 1). 

To convert:  $\lambda^{\rm exp} = \lambda^{\rm Poiss}/{\rm (time\ window)}$. If the time window for the Poisson process is of unit length, then $\lambda^{\rm exp} = \lambda^{\rm Poiss}$.

If you want to read more about the Poisson process and the rate, here is a [great blog post](https://towardsdatascience.com/the-poisson-distribution-and-poisson-process-explained-4e2cb17d459) with many good examples. 

#### Example 2: Waiting for a Shared Resource

Suppose $X$ denotes the duration of a telephone call and is exponentially distributed with rate parameter $\lambda = \frac{1}{180}$ per second. Suppose you arrive at a telephone booth to find that  there is person already  using it. What is the probability that you have to wait more than 50 seconds before you get to use the booth?

**Solution:**  
First  even though the person was already using the phone booth when you arrived, the remaining time $Y$ which the person will use the booth is still $\textbf{Expo}(\lambda)$. This follows from the memory-less property.  Thus,

\begin{eqnarray*}
F_Y(t) &=& P(Y \leq t) \\[2mm]
            &=& \int_{-\infty}^{t} f_X (x) dx \\[2mm]
            &=& 1 - e^{- \lambda t}
\end{eqnarray*}

We need to find $P(Y > 50)$ which is given by

\begin{eqnarray*}
P(Y > 50) &=& 1 - P(Y \leq 50) \\[2mm]
            &=& 1 - F_Y(50) \\[2mm]
            &=& 1 - (1 - e^{- \lambda 50})  \\[2mm]
            &=& e^{- \frac{50}{180} }
\end{eqnarray*}


#### Example 3: CPUs and Job Queues  (waiting if there are multiple resources) 
Two cores in a CPU  serve jobs from a single queue. Core 2 is more powerful than Core 1. The time it takes Core 1 to serve a randomly selected job is distributed as $\text{Expo}(\frac{1}{5})$ so that the average service time, $E(X)  = 1/\lambda$, is 5 minutes. The other, more powerful core's service times are distributed as $\text{Expo}(\frac{1}{4})$, averaging 4 minutes. The cores work independently. Both cores are busy at the moment a new job arrives in the queue. What is the probability it will be longer than 5 minutes before the new job starts to be served?


![two-core-problem](Figures/Two-Core-Problem.jpg)


**Solution:**
Let $X_1$ and $X_2$ be the random variables that denote the execution times in core 1 and core 2, respectively. We are given that $X_1 \sim \text{Expo} (\frac{1}{5})$ and  $X_2 \sim \text{Expo} (\frac{1}{4})$. Let $W$ denote the time that  the new job must wait before it starts to be served.  Since the new job will start to be served as soon as any one of the current jobs is finished, we can see that $W = \min(X_1 , X_2)$. We want to $P(W > 5)$


\begin{eqnarray*}
  P(W > 5) &=& P(\min(X_1 , X_2) > 5) \\[2mm]
           &=& P(X_1 > 5 \cap X_2 > 5) \\[2mm]
           &=& \underbrace{P(X_1 > 5) \times P(X_2 > 5)}_{\text{due to independence}} \\[2mm]
           &=&  e^{-\frac{1}{5} \times 5} e^{-\frac{1}{4} \times 5} \\[2mm]
           &=& e^{-\frac{9}{4}}
\end{eqnarray*}

## VII.6. Normal Distribution 

This is an important distribution. By  the Central Limit Theorem (discussed in detail later), the sum of lots of independent and identically distributed (iid) random variables looks like a Normal distribution.  If $X$ has a Normal distribution, the we write 

$$X ~\sim \text{Norm}(\mu, \sigma^2) = {\rm N}(\mu, \sigma^2)$$ 

where $\mu$ and $\sigma^2$ are the two parameters. The pdf of the distribution is given by

\begin{equation}
    	f_X (x) = \frac{1}{\sqrt{2 \pi} \sigma} e^{- \frac{(x - \mu)^2}{2 \sigma^2}} ~~~~~~~~ - \infty < x < \infty
\end{equation}

The normal distribution is known as the Gaussian distribution.  In the popular literature it is called "the Bell curve". We look at the PDF and the CDF next respectively:

```r
mean=70
sd=15
x   <- seq(10, 130, 1) # Generate a sequence of integers from 10 to 130 for x values
y   <- dnorm(x,mean, sd)  # Calculate fx each different x value
plot(x,y, type="l", main = "PDF of Normal Distribution", xlim=c(10,130), ylim=c(0,max(0.028,(1/sqrt(2*3.14*sd^2)))),ylab = "PDF")
abline(v=mean)  #add a vertical line at the mean value 
```

```r
mean=70
sd =15
x   <- seq(10, 130, 1) # Generate a sequence of integers from 10 to 130 for x values
y   <- pnorm(x,mean,sd)  # Calculate the CDF of x
plot(x,y, type="l", main = "CDF of Normal Distribution", ylab = "CDF")
abline(v=70)  #add a vertical line at the mean value 
Notabline(h=0.5) #add a horizontal line for CDF=0.5 
```

Note from the plot above, you can see that the median equals the mean for a normal distribution. 

### VII.6.a) Properties
The following are some of the properties of Normal random variable

1. The pdf is:
\begin{equation}
    	f_X (x) = \frac{1}{\sqrt{2 \pi} \sigma} e^{- \frac{(x - \mu)^2}{2 \sigma^2}} ~~~~~~~~ - \infty < x < \infty
\end{equation}


2. The pdf is symmetrical around  $\mu$. If we consider two values $x_1 = \mu + t$ and $x_2 = \mu - t$ then you can show that $f_X(x_1) = f_X(x_2)$.


2. The maximum value of $f_X (x)$ occurs at $x = \mu$ and the maximum value is $\frac{1}{\sqrt{(2 \pi) \sigma}}$.

3. From the maximum value the function decays as $e^{-x^2}$ (very fast).


4. Expectation:   $E(X) = \mu$ 


5. Variance:     $Var(X) = \sigma^2$

Now you see that the parameters of the density function, Norm($\mu,\sigma^2$), are the mean and the variance of the distribution.


### Facts
The following are some additional facts about the Normal distribution. We will not prove them. Some of them are easy to prove while some are not.

1. $f_X (x)$ is a valid pdf. In order to prove this we need to show that a) $f_X (x) \geq 0$ and b) $\int_{-\infty}^{\infty} f_X (x) = 1$.


2. $F_X (a) = P(X \leq x) = \int_{-\infty}^{a} f_X (x) dx$. This is by definition.


3. If $X \sim N(\mu, \sigma^2)$ and suppose $Y = b + aX$, where $a$ and $b$ are constants, then $Y \sim N(a \mu + b, a^2 \sigma^2)$.  In other words a linear function of $X \sim N(\mu, \sigma^2)$ is also has a normal distribution but with the new mean of $a \mu + b$ and new variance of $a^2 \sigma^2$.




### VII.6.b) The Standard Normal Distribution $Z \sim N(0,1)$


There is a **special Normal distribution $Z \sim N(0,1)$ which is called the Standard Normal**. This is a Normal distribution with mean 0 and variance 1.

```r
x   <- seq(-3, 3, 0.01) # Generate a sequence of integers from -3 to 3
y   <- dnorm(x,mean=0, sd=1)  # Calculate the PDF of x
plot(x,y, type="l", main = "PDF of Standard Normal Distribution", ylab = "PDF")
abline(v=0)
```

```r
x   <- seq(-3, 3, 0.01) # Generate a sequence of integers from -3 to 3
y   <- pnorm(x,mean=0, sd=1)  # Calculate the CDF of x
plot(x,y, type="l", main = "CDF of Standard Normal Distribution", ylab = "CDF")
abline(v=0)
abline(h=0.5)
```

We can compute the probabilities for any normal distribution in terms of the standard normal distribution as shown here: 

1. We will use $\Phi(x)$ to denote $P(Z \leq x)$. Notice that because of the symmetry property

\begin{equation}
\Phi(x) = 1 - \Phi(-x)
\end{equation}

<img src="Figures/Z-score-reflect.png" width="600">
<center>From the image it is easy to see that $\Phi(-x) + \Phi(x)=1$. </center>



This is useful since the likelihood of events can be stored in a Standard Normal table (see figure below). Due to the symmetry, the table only needs to include the positive values of $x$.

2. If $X ~\sim N(\mu, \sigma^2)$, we can write it as $X = \mu + \sigma Z$. Using Fact 3 above with $b=\mu$ and $a=\sigma$ we find 

$$X \sim N(\sigma \cdot \mu_{N(0,1)} + \mu, \sigma^2 \cdot \sigma^2_{N(0,1)}) = N(\mu, \sigma^2)$$ 

where the last equality uses the fact that $\mu_{N(0,1)}=0$ and $\sigma^2_{N(0,1)}=1$. Thus 

$$\frac{X - \mu}{\sigma} = Z ~\sim N(0,1)$$
            
            
3. We can use the above fact to show how to use $\Phi(x)$ to compute the CDF for any Normal distribution $X ~\sim N(\mu, \sigma^2)$.

\begin{eqnarray*}
	           F_X (a) &=& P(X \leq a) \\[2mm]
	                   &=& P\left(\frac{X - \mu}{\sigma} \leq \frac{a - \mu}{\sigma}\right) \ \ {\rm transformed\ both\ variables\ into\ standard\ normal\ values}\\[2mm]
	                   &=& P\left(Z \leq \frac{a - \mu}{\sigma}\right) \\[2mm]
	                   &=& \Phi\left(\frac{a - \mu}{\sigma}\right)
\end{eqnarray*}


### VII.6.c) Z Table and "Z-score"

The values of $\Phi(Z)$ are tabulated for all values of Z in increments of 0.01 reading from left to right and then moving down the rows. So given a value of Z you can read off $\Phi(Z)$.  But you can also do the opposite. If you know the numerical value of $\Phi(Z)$ you can find where it appears in the table and read off the corresponding value of $Z$. #### Example 1: More Realistic Batch Job Run Times

Suppose the runtime of a job on the CORI supercomputer is a random variable $X$ which has a Normal distribution with mean $\mu = 3$ mins  and standard deviation $\sigma = 3$ mins, i.e., $\sigma^2 = 9$. Use the Z-scores to find the probability that the run time of the job will be between 2 and 5 mins.


**Solution:**  

We have to map all the values into $N(0,1)$.  So $Z = \frac{X-\mu}{\sigma}$, likewise the upper bound is converted $\frac{5 - \mu}{\sigma}$ and the lower bound $\frac{2-\mu}{\sigma}$. 

\begin{eqnarray*}
	P(2 \leq X \leq 5) &=& P(\frac{2 - \mu}{\sigma} \leq \underbrace{\frac{X- \mu}{\sigma}}_{\text{This is $Z$}} \leq \frac{5 - \mu}{\sigma}) \\[3mm]
	                   &=& P(\frac{2 - 3}{3} \leq Z \leq \frac{5 - 3}{3}) \\[2mm]
	                   &=& P(\frac{-1}{3} \leq Z \leq \frac{2}{3}) \\[2mm]
	                   &=& \Phi(\frac{2}{3}) - \Phi(\frac{-1}{3}) \ \ \ \ \ \ \ \ \ \  {\rm recall}\ \ P(a < X \le b) = F_X(b) - F_X(a)\\[2mm]
	                   &=& \Phi(\frac{2}{3}) - (1 - \Phi(\frac{1}{3})) \ \ \ \ \ \  {\rm recall} \ \ \Phi(-x) = 1 - \Phi(x)\\[2mm]
	                   &=& \Phi(\frac{2}{3}) + \Phi(\frac{1}{3}) - 1 \\[2mm]
	                   &=& 0.378
\end{eqnarray*}

We can read the approximate values from the table.  

- $\Phi(\frac{2}{3}) \approx \Phi(0.66) = 0.7454$


- $\Phi(\frac{1}{3}) \approx \Phi(0.33) = 0.6293$


- $\Phi(\frac{2}{3}) + \Phi(\frac{1}{3}) - 1 \approx 0.3747$


We can also get the exact values from R using the pnorm function:  


```r
pnorm((2/3),0,1) + pnorm((1/3),0,1) -1 
```

Suppose the execution times of jobs in  the CORI supercomputer are approximately normally distributed. If 25\% of the jobs take less than 180 seconds (s) and 25\% take more than 325s, approximately what fraction take

1. less than 200s?
2. between 280s and 320s?

**Solution:**  
Let $X$ be the random variable that denotes the execution time of a job. $X \sim N(\mu, \sigma^2)$. We do not know what is $\mu$ and $\sigma^2$ and we have to find them from the information provided.
Since 25\% of the jobs take less than 180s, we have

\begin{eqnarray*}
	P(X < 180) &=& 0.25 \\[2mm]
	P(\frac{X - \mu}{\sigma} < \frac{180 - \mu}{\sigma}) &=& 0.25 \\[2mm]
	P(Z < \frac{180 - \mu}{\sigma}) &=& 0.25 \\[2mm]
	\Phi(\frac{180 - \mu}{\sigma}) &=& 0.25 \\[2mm]
	1 - \Phi(\frac{\mu - 180}{\sigma}) &=& 0.25 ~~~~~~~~ \text{Used the fact that $\Phi(x) = 1 - \Phi(-x)$} \\[2mm]
	 \Phi(\frac{\mu - 180}{\sigma}) &=& 0.75 \\[2mm]
	\frac{\mu - 180}{\sigma} &=& 0.675 ~~~~~~~~~~~~\text{Used the Z table and read the value of $x$ such that $\Phi(x) = 0.75$}\\
\end{eqnarray*}


We are also given that 25\% of the jobs take greater than 325s from which we have

\begin{eqnarray*}
	P(X > 325) &=& 0.25 \\[2mm]
	1 - P(X < 325) &=& 0.25 \\[2mm]
	P(X < 325) &=& 0.75 \\[2mm]
	P(\frac{X - \mu}{\sigma} <\frac{325 - \mu}{\sigma}) &=& 0.75 \\[2mm]
	P(Z < \frac{325 - \mu}{\sigma}) &=& 0.75 \\[2mm]
	\Phi(\frac{325 - \mu}{\sigma}) &=& 0.75 \\[2mm]
	\frac{325 - \mu}{\sigma} &=& 0.675 ~~~~~~~~~~~~\text{Used Z table and read the value of $x$ such that $\Phi(x) = 0.75$}\\
\end{eqnarray*}

Now we have two equations and two unknowns $\mu$ and $\sigma^2$ which we can solve to find $\mu = 252.5$ and $\sigma =  107.4074$.




## VII.7. Power-law distribution 

The normal distribution describes many phenomena in the world around us (and we will later use the Central Limit Theorem to show why).  But there are many important phenomena that are instead described by a power law distribution.  

The pdf is

$$f_X(x) = a x^{-k}$$  

We must have $k \ge 1$ to satisfy normalization, and $a$ is a normalization constant that depends on the value of $k$.  Taking the log of both sides we get a straight line with negative slope $-ak$: 

$$\ln f_X(x) = - ak \ln x$$

Here is an example using data on the distribution of cities sizes, which follows a power law. On the left it is shown in a linear-linear scale plot, and on the right on a log-log scale plot showing the straight line behavior (the deviations are from the data sampling errors). 

<img src="Figures/PowerLaw-cities.png" width="700">



Random variables that follow a power law distribution include:

- Distribution of income,
- Distribution of wealth,
- Magnitude of earthquakes,
- Number of friends on Facebook,
- Size of cities according to population,
- Trading volumes on the stock market,
- Word frequencies,
- Number of hits on web sites,
- Number of links a web site has,
- Number of connections in protein interaction networks,
- Number of connections in gene regulatory networks. 


Power laws are called "fat-tailed distributions", or "heavy-tailed distribution" because a large fraction of the probability mass is found in the largest values of X.  It means that the typical value of $X$ is small, but that there are massive values of $X$ with significant probability. Here is how the far-right-tail events of multiple distributions compare:

## VII.8.  Entropy

Claude E. Shannon published "A Mathematical Theory of Communication" in the Bell System Technical Journal in 1948. With this groundbreaking work came the dawn of the information age. Here we briefly mention just a small subset of this work. We begin with the concept of Information Entropy. About a decade before Shannon's work, Morse Code had been widely proliferated. In Morse Code, the letter E is represented with just a single dot $\cdot$, while the letter Y is represented by -- $\cdot$ -- --. Why is this?

Indeed E is quite common, while Y is rare. Therefore, it makes practical sense to represent the E with a shorter *codeword*. A codeword is just an alternative method of representing information for transmission over some channel. To form wordds, letters in the English alphabet are chosen with some probability, so they can be modeled with a stochastic process. Can we come up with some formal way to choose our codewords optimally? It is precisely this problem Claude Shannon was concerned with. The formalization of this problem is Shannon Entropy.

Suppose a random variable $X$ takes on values $x_1, x_2,  \ldots, x_k$ with probabilities $p_1, p_2, \ldots, p_k$. We define entropy $H(X)$ as follows

\begin{equation}
H(X) = - \sum_{i=1}^{n} p_i log_2(p_i)
\end{equation}

The log is taken with base 2, and outputs the amount of information contained in the stochastic process in bits. $H(X)$ is also called the Shannon entropy and is the foundation of information theory.

The entropy is essentially the uncertainty that we have about the outcome of a random variable.  A distribution like the uniform distribution has the maximum entropy since all values are equally likely.  A sharply peaked distribution, for instance a normal distribution with $\sigma=1$, has very low entropy since $X$ is mostly likely to take on the value $\mu$. 

Let's look at these two examples, first a uniform distribution, then the normal distribution. 

```r
a=10
b=20

x = seq(a,b,1)
prob = dunif(x, a, b)
entropy = -sum(prob * log2(prob))
plot(x,prob,ylim=c(0.0,0.14))
print(entropy)
```

```r
mean=70
sd=15
x   <- seq(10, 130, 1) # Generate a sequence of integers from 10 to 130 for x values
prob   <- dnorm(x, mean, sd)  # Calculate fx each different x value
plot(x, prob, type="l", main = "PDF of Normal Distribution", ylim=c(0,max(0.028,(1/sqrt(2*3.14*sd^2)))), ylab = "PDF")
entropy = -sum(prob * log2(prob))
print(entropy)
```

Entropy forms the cornerstone of information theory.  It also forms the cornerstone of thermodynamics, telling us that in a spontaneous process entropy increases until it is maximized. (e.g., think of your dorm room and what would happen if you left it alone for a month --- papers would fall on the floor, dust would coat every surface -- it would be more disordered, meaning that the entropy increased.)
