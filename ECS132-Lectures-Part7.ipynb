## X.A. Inequalities

These are two inequalites that can be used as bounds on the values of probabilities. 

### Markov Inequality

If $X$ is a random variable that takes only *nonnegative values*, then for any $a > 0$

\begin{equation*}
        P(X \geq a) \leq \frac{E(X)}{a}
      \end{equation*}


This provides a bound on the probability when only the mean of the distribution is known. 


#### Example 1

Suppose the number of e-mails received in a day is a random variable $X$ with mean 25. What is the probability that today the number of e-mails will exceed 30 (greater than or equal to 30)? 

By Markov inequality 
\begin{eqnarray*}
P(X \geq 30) & \leq & \frac{E(X)}{30} \\
          & \leq & \frac{25}{30} \\
          & \leq & \frac{5}{6}
\end{eqnarray*}

### Chebyshev's  Inequality

If $X$ is a random variable with finite mean $\mu$ and variance $\sigma^2$, then for any value $k >0$
      \begin{equation*}
        P(|X - \mu | \geq k) \leq \frac{\sigma^2}{k^2}
      \end{equation*}


This provides a bound on the probability when only the mean and the variance of the distribution is known. 

#### Example 1

If the variance $\sigma^2$ of the number of e-mails received is 10, what is the probability that today the number of e-mails received is between 20 and 30? 

By Chebyshev's  Inequality

\begin{eqnarray*} 
P(|X - 25| \geq 5) & \leq & \frac{\sigma^2}{5^2} \\
                   & \leq & \frac{10}{25} \\
                   & \leq & \frac{2}{5}
\end{eqnarray*}

Thus, $P(|X - 25| <  5) = 1 - P(|X - 25| \geq 5) \geq 1 - 2/5 \geq 3/5$


### Weak law of large numbers

Let $X_1 , X_2, \ldots$ be a sequence of independent and identically distributed (IID) random variables with finite mean $E(X_i) = \mu$. Then for any $\varepsilon > 0$,

\begin{equation*}
        P \left(\ \left|\frac{X_1 + X_2 + \ldots + X_n }{n} - \mu \ \right| \geq \varepsilon \right) \rightarrow 0 ~~ \text{as} ~~ n \rightarrow \infty
\end{equation*}

In other words, the probability that the value of the sample mean of the $X_i$'s is greater than distance $\epsilon$ from the mean of the real underlying distribution $\mu$ goes to zero as the number of samples goes to infinity. 

This result is true even if $X_i$'s  have finite mean but infinite variance. For the finite variance case the result follows from Chebyshev's  Inequality. 


## X.B. Sample Mean and Sample Variance

When we have a large population we cannot measure a random variable $X$ for each member of the population. Instead we often have a limited number of **observations**. Likewise, when you are running your numerical simulations in R to model the probability of random variables, like in the coupon collector problem or the birthday problem, you generate a set of observations and from those make conclusions about the underlying distribution which generated the random variable. 

**The goal of statistical inference (estimation) is to infer (estimate)  characteristics (parameters) when we cannot observe the entire population or to infer (estimate) the underlying distribution that generated the observed data.** 

Consider making observations from a population of individuals: 

1. Example 1: Suppose 120 students in the class are classified according to their blood type. The population is the entire class. It this case it can be possible to observed (i.e., sample) the enter population.


2. Example 2: Suppose we would like to classify all 18 year olds in California according to their blood type. The population of all the 18 year olds in California is in the millions so it is infeasible that every member of the population can be observed. 


**Observations** are values of a random variable $X$ sampled from the density function $f_X(x)$. 


When sampling a subset of the population the goal is to obtain a  sample that is representative of the population. It is  important that sampling is done randomly without any bias. If the sampling is biased then inferences can underestimate or overestimate the characteristics of the population. 

Assume we have a total of $n$ observations in our sample. Let $X_i, i = 1,2, \ldots, n$ be the random variable that represents the $i^{th}$ sample each having the same probability density function $f_X (x)$.  If $X_i$'s are random (independent) samples then 

\begin{equation}
f(x_1 \cap x_2 \cap \ldots \cap x_n) = f_X (x_1) f_X (x_2) \ldots f_X (x_n)
\end{equation}

### X.B.1. Sample Mean $\overline{X}$

  \begin{equation*}
    \overline{X} = \frac{X_1 + X_2 + \ldots + X_n }{n}
  \end{equation*}
  
If $x_1, x_2, \ldots x_n$ are the sample values (i.e., the values assumed by the random variables $X_1, X_2, \ldots X_n$, respectively), $\overline{x} = \frac{1}{n}\sum_{i=1}^n x_i$ is the value assumed by the sample mean $\overline{X}$. 

#### Example 1

Suppose packet sizes (in bytes) of 5 randomly sampled packets are 64, 256, 238, 750, and 64. Then $\overline{x} = (64 + 256 + 238 + 750 + 64)/5 = 274.2$. 

### X.B.2. Sample variance $S^2$

  \begin{equation*}
    S^2 = \frac{1}{n-1} \sum_{i=1}^{n} \left(X_i - \overline{X}\right)^2
  \end{equation*}


#### Example 2

Suppose we get sample inter-packet delays (IPDs) in microseconds  from 4 routers and they are 12, 15, 17, and 20 microseconds. Then we can easily calculate that 
$\overline{x} = (12+15+17+20)/4 = 16$ and hence 

\begin{eqnarray*}
S^2 &=& \frac{1}{3} \sum_{i=1}^{4}\left(x_i - 16\right)^2 \\[3mm]
    &=& \frac{34}{3}
\end{eqnarray*}


Some notes: 

1. Sample Mean and Sample Variace are random variables. Every sample will give a different value of the sample mean and sample variance. 


2. For the sample variance we divide by $n-1$. This makes $S^2$ an unbiased estimator of $\sigma^2$. 


### More on why the sample variance $S^2$ is calculated by dividing by $n -1$ and not $n$? 

First note in computing the sample variance $S^2$  we finding the sum of square of the residuals between the sample and the sample mean.  Specifically, the residuals 
are $x_1 - \overline{x}, x_2 - \overline{x}, \ldots , x_n - \overline{x}$, where $x_i, i = 1, \ldots n $ are the $n$ samples and the value of the sample mean $\overline{x} = \frac{1}{n}\sum_{i=1}^n x_i$. 

Note that while there are $n$ independent samples $x_i, i = 1, \ldots n $, there are only $n-1$ independent residuals since the sum the sum of the residuals is equal to 0.  


### More formal definition

Sample mean $\overline{X}$ and Sample variance $S^2$ are random variables. They and any function of these are called statistics. A statistic $\Theta$ is said to be an unbiased estimator of the parameter $\theta$ if

\begin{equation} 
E(\Theta) = \theta
\end{equation} 

For $S^2$ to be an unbiased estimator of population variance $\sigma^2$ it must be divided by $n - 1$.  In other words, if we had used $n$ in the denominator then $E(S^2) \neq \sigma^2$. 

### X.B.3 Error bars on plots 

The sample mean is typically reported in plots (as we have done now several times during the numerical simulation problems on the homework assignments). But what is missing in our plots are **error bars**, in other words, communicating how much uncertainty there is in your estimate obtained from the sample mean. 

Error bars are drawn above and below a data point, which extend the length of the **standard error** in each direction. The standard error is equal to the square-root of the sample variance divided by the number of observations: 

$$\sqrt{\frac{S^2}{n}} = \frac{S}{\sqrt{n}}$$ 

where $n$ denotes the number of observations.

## X.C. Central Limit Theorem

Central limit theorem (CLT): Let $X_1 , X_2, \ldots X_n$ be a sequence of independent and identically distributed (IID) random variables with mean $\mu$ and variance $\sigma^2$. It does not matter what the specific distribution is, only that it has mean $\mu$ and variance $\sigma^2$. Then the distribution of Sample Mean $\overline{X}$ approaches  $N(\mu, \frac{\sigma^2}{n})$ when  $n$ is large. Thus

\begin{equation*}
\frac{X_1 + X_2 + \ldots + X_n}{n} \sim N\left(\mu, \frac{\sigma^2}{n} \right)
 \end{equation*}
 when $n$ is large. 


Using the linear transformation to map onto a standard normal distribution, we can show that

\begin{equation*}
        \frac{X_1 + X_2 + \ldots + X_n  - n\mu}{\sigma \sqrt{n}}
\end{equation*}

tends to the standard Normal $Z \sim N(0,1)$ when  $n$ is large.

The Old Faithful Data set on eruption duration did not follow any particlar parametric distribution, but we will see it still obeys the Central Limit Theorem (CLT). #### Example 1
Suppose the number of accesses to a website by any randomly selected customer has a mean value 3.2 and standard deviation 2.4. Among 100 randomly slected customers, how likely is the sample mean of the number of times the website is accessed to exceed 4.0? 

##### Answer 

Let $X_i$ be the number of times the $i$th customer accesses the website.  The sample mean of the number of times the website is accessed by 100 customers is given by 

\begin{equation}
 \overline{X} = \frac{X_1 + X_2 + \ldots + X_{100}}{100}
\end{equation}

We need to find $P(\overline{X} > 4)$ which is given by 

\begin{eqnarray*}
 P(\bar{X} > 4) &=& P(\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}} > \frac{4 - \mu}{\frac{\sigma}{\sqrt{n}}}) \\[3mm]
                &=& P(Z > \frac{4 - 3.2}{\frac{2.4}{\sqrt{100}}}) = 1 - P(Z \le \frac{4 - 3.2}{\frac{2.4}{\sqrt{100}}}) \\[3mm]
                &=& 1 - \Phi(3.33) 
\end{eqnarray*}

#### Example 2

An astronomer is interested in measuring the distance, in light-years, from her observatory to a distant star. Although the astronomer
has a specific measuring protocol, she knows that because of changing atmospheric conditions and normal error, each time a measurement is made, it will not yield the exact distance, but merely an estimate. As a result, the astronomer plans to make a series of measurements and then use the average value of these measurements as the  estimated value of the actual distance. If the astronomer believes that the value of the measurements are independent and identically distributed (IID) random variables having a common mean $d$ (the actual distance) and a common variance of 4, how many measurements should she have to make to be reasonably sure (say with 95\% probability) that her estimated distance is accurate to within 0.5 (postive and negative)  light-years?


##### Answer

We are told $\mu=d$ light-years and $\sigma^2=4$. Suppose that the astronomer decides to make $n$ observations. If $X_1, X_2, \ldots, X_n$ are the $n$ measurements, then, from the Central Limit Theorem, it follows that

\begin{equation*}
Z = \frac{\overline{X} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{\overline{X} - d}{\frac{2}{\sqrt{n}}} \sim N(0,1)
\end{equation*}

We want to find $P( d-0.5 \leq  \overline{X}   \leq d+0.5)$.  Transforming to a standard normal distribution: 

$$P\left(\frac{(d-0.5)-d}{\frac{2}{\sqrt{n}}} < \frac{X-d}{\frac{2}{\sqrt{n}}} < \frac{(d+0.5)-d}{\frac{2}{\sqrt{n}}}\right) $$

Simplifying the expression:

$$ P\left(- \frac{\sqrt{n}}{2} 0.5 < Z <  \frac{\sqrt{n}}{2} 0.5\right)$$

Using these facts: 

\begin{eqnarray*}
P( d-0.5 \leq  \overline{X}   \leq d+0.5) &=& P(-0.5\frac{\sqrt{n}}{2}  \leq Z \leq 0.5\frac{\sqrt{n}}{2}) \\
                                                                                               &=& \Phi(\frac{\sqrt{n}}{4}) - \Phi(- \frac{\sqrt{n}}{4}) \\
                                                                                               &=& 2 \Phi(\frac{\sqrt{n}}{4}) - 1 \\
\end{eqnarray*}

We want to find $n = n^*$ such that

\begin{eqnarray*}
          2 \Phi(\frac{\sqrt{n^*}}{4}) - 1   = 0.95 \\
          \Phi(\frac{\sqrt{n^*}}{4}) = 1.95/2 = 0.975
\end{eqnarray*}

We can now use the Standard Normal table  or qnorm to get  

\begin{equation*}
\frac{\sqrt{n^*}}{4} = 1.96
\end{equation*}

from which we get $n^* \approx 62$. The astronomer should make 62 measurements. 


```r
qnorm(0.975, 0, 1)
```

## XI. Maximum Likelihood Estimation

Given a sample, that we know follows a specific distribution, how do we estimate the parameter of the distrution. For instance suppose we know a random variable is distributed according to a Poisson distribution but we do not know the parameter $\lambda$. 

First, let's consider a concrete example. Suppose we are in a factory where each part can be defective or good (non-defective) but we can only sample $n$ parts. How do estimate the probability of defective part for the full population? 

Let $X_1, X_2, \ldots, X_n$ be independent random variables that are taken from a distribution with density fiunction $f_X(x; \theta)$ where $\theta$ is the parameter of the distribution. We define likelihood function $L(x_1,x_2, \ldots, x_n; \theta)$ which is given by 

\begin{eqnarray*} 
L(x_1,x_2, \ldots x_n; \theta) &=& f(x_1 \cap x_2 \cap \ldots \cap x_n; \theta) \\
                               &=& f(x_1; \theta) f(x_2; \theta) \ldots f(x_n; \theta) 
\end{eqnarray*}

where $x_1, x_2, \ldots, x_n$ sample values, i.e., the observed values of $X_1, X_2, \ldots, X_n$. 

The maximum likelihood estimator is the value of the parameter $\theta$  that results in the maximum value of the joint probability. Recall that to find a maximum of a function you take the first derivative with respect to $\theta$ and set it equal to zero 

$$\frac{dL}{d\theta}=0$$

and solve the equation for $\theta$. You can show the value is a maximum and not a minimum by taking the second derivative with respect to $\theta$ and showing it is negative

$$\frac{d^2L}{d\theta^2}< 0$$


### Example 1

3 items are selected  from a population with unknown non-defective (ND) probability $p$. Whether each sample is non-defective is an independent Bernouli trial with probability $p$. Suppose for the three samples we get $x_1 = {\rm ND}, x_2 = {\rm ND}, x_3 = {\rm D}$.  The likelihood function $L$ is given by 

\begin{eqnarray*}
L & = & f(x_1; p) f(x_2; p) f(x_3; p) \\[2mm]
  & = & p(ND; p) \times p(ND; p) \times p(D; p) \\[2mm]
  & = & p \times p \times (1-p)  \\[2mm]
  & = & p^2 (1 - p) \\[2mm]
  & = & p^2 - p^3 
\end{eqnarray*}

We want to find the value of the parameter $p$ that maximizes $L$. By differentiating $L$  with respect to $p$ and equating to 0 we can show that $p^* = 0.6666$ maximizes $L$. To show that it is indeed the maximum, take the second derivative and show that it is negative. Doing the first step explicitly:

\begin{eqnarray*}
\frac{dL}{dp} &=& 2p - 3p^2\\
&=& p(2-3p) = 0.
\end{eqnarray*}

Rearranging we find $p=2/3$. 

Taking the second derivative

\begin{eqnarray*}
\frac{d^2L}{dp^2} &=& 2 - 6p
\end{eqnarray*}

Evaluating this at $p=2/3$ we find $\frac{d^2L}{dp^2} = -2$ so the maximum of $L$ occurs at $p=2/3$.

## Log likelihood

In most cases it is easier to work with log of the likelihood function and we find the value of the parameter that maximizes the log likelihood function. This is shown in the example below. 

### Example 2 

Suppose 10 late stage cancer subjects   are injected with  a cancer drug that is designed to increase their survival rate. Unfortunately, they all do ultimately succumb to cancer. The survival times, in months, are 14, 17, 27, 18, 12, 8, 22, 13, 19, and 12. Assume that survival times are exponential distributed. Give a maximum likelihood estimate of the mean survival time.

#### Answer

We know that the probability density function for the exponential random variable X is given by

\begin{equation*}
% \nonumber to remove numbering (before each equation)
  f_X (x; \lambda) = {\lambda} e^{- {x}{\lambda}}     ~~~~ x > 0, \lambda > 0
\end{equation*}

where $\lambda$ is the rate and hence $\frac{1}{\lambda}$ is the mean survival  time. We define the likelihood function

\begin{eqnarray*}
  L(x_1, x_2, \ldots, x_n : \lambda) &=& f_X (x_1 \cap x_2 \cap \ldots \cap x_n ; \lambda) \\[3mm]
                                   &=& f_X (x_1 ; \lambda)f_X (x_2 ; \lambda) \ldots f_X (x_n ; \lambda) \\[3mm]
                                   &=& {\lambda^n} e^{-\lambda (x_1 + x_2 + \ldots + x_n )}
\end{eqnarray*}

Now we take the log of the Likelihood function which gives us

\begin{eqnarray*}
  \ln(L(x_1, x_2, \ldots, x_n : \lambda)) &=& \ln({\lambda^n} e^{-\lambda (x_1 + x_2 + \ldots + x_n )})\\[3mm]
                                         &=& n \ln \lambda  -  \lambda (x_1 + x_2 + \ldots + x_n )
\end{eqnarray*}

Now differentiating $\ln(L(x_1, x_2, \ldots, x_n : \lambda))$ with respect to $\lambda$ and setting it to 0 we obtain the following

\begin{eqnarray*}
  \frac{d\ln(L(x_1, x_2, \ldots, x_n : \lambda))}{d\lambda} &=& n \frac{1}{\lambda} - (x_1 + x_2 + \ldots + x_n ) = 0
\end{eqnarray*}

Rearranging we find:

\begin{eqnarray*}
                                    \lambda &=& \frac{n}{x_1 + x_2 + \ldots + x_n} \\[3mm]
                                            &=& \frac{10}{14 + 17 + 27 + 18 + 12 + 8 + 22 + 13 + 19 + 12}\\[3mm]
                                            &=& \frac{10}{162} \\[3mm]
                                            &=& 0.0617284
\end{eqnarray*}

The mean survival time is $\frac{1}{\lambda}$ which is 16.2 months.

## XII.A. Linear model -- a model of Linear Correlation 

We consider two independent, random variables $X$ and $Z$ with $X \sim N(0, \sigma_x)$ and $Z \sim N(0, \sigma_z)$. We define 

$$Y = mX + Z$$ 

and want a measure of correlation between $X$ and $Y$.  We look at three example cases one-by-one below. The cases are:

1. Perfect correlation ($\sigma_z$ is zero): for instance $\sigma_z = 0$, $\sigma_x = 1$


2. Some correlation ($\sigma_z$ is small compared to $\sigma_x$): for instance $\sigma_z=0.25$, $\sigma_x=1$ 


3. Zero Correlation ($m=0$): In this case $X$ and $Y$ are independent. 

Looking at each case in detail: 

1. Perfect Correlation: If $\sigma_z = 0$, $\sigma_x = 1$  we have perfect correlation. The following R code and figure show this.

```r
X = rnorm(1000, 0, 1)
Z = rnorm(1000, 0, 0)
Y = 0.5*X + Z
plot(X,Y, main="Perfect Correlation")
```

2. Some Correlation: In this case we set $\sigma_z > 0$. The following code is used to generate the figure which shows some correlation between $X$ and $Y$

```r
X = rnorm(1000, 0, 1)
Z = rnorm(1000, 0, 0.25)
Y = 0.5*X + Z
plot(X,Y, main="Some Correlation")
```

3. Zero Correlation: In this case we set $m=0$.  Recall we defined $Y=mX + Z$ so if $m=0$ then $X$ and $Y$ are independent. 

```r
X = rnorm(1000, 0, 1)
Z = rnorm(1000, 0, 0.25)
Y = 0.0*X + Z
plot(X,Y, main="Zero Correlaton")
```

## XII.B Measure of Correlation

We are looking at the linear relationship $Y=mX + Z$. We define the correlation coefficient $r$ between $X$ and $Y$ as as follows

\begin{eqnarray*}
  r &=& \frac{m \sigma_x}{\sigma_y} = \frac{m \sigma_x}{\sqrt{m^2 \sigma_x^2 + \sigma_z^2}} \\
\end{eqnarray*}

### Key Points

The following are some of the key points

1. The correlation coefficient $r$ measures the proportion of the standard deviation of $Y$ that is contributed by $mX$. If this proportion is large, then the correlation between $X$ and $Y$ will be large. In the extreme case when $\sigma_z = 0$, then this proportion $r = 1$ and $X$ and $Y$ are perfectly correlated. On the other hand if $m = 0$, then $X$ and $Y$ are not correlated and hence $r = 0$.


2. If $m < 0$ then $X$ and $Y$ will be negative correlated. The value of $-1 \leq r \leq 1$.


3. If $X$ is a random variable with variance $\sigma^2$, then $Var(aX) = a^2 \sigma^2$ where $a$ is some constant scaling factor. Also, if $X$ and $Y$ are independent random variables with variance $\sigma_x^2$ and $\sigma_y^2$, respectively, then $Var(aX + bY) =  a^2 \sigma_x^2 + b^2 \sigma_y^2$.





### Determining Correlation Coefficient Given a Set of Data Points 

In general we will be given a set of data points and we need to determine the correlation coefficient $r$.   This is determined in terms of the covariance.

## XII.C Measure of Covariance 

The covariance of two random variables $X$ and $Y$ is defined as 

\begin{equation*}
  Cov(X,Y) = E\left[(X - \mu_x)(Y - \mu_y)\right]
\end{equation*}

where $\mu_x$ and $\mu_y$ are the mean of $X$ and $Y$.  If $X = Y$, then 

$$Cov(X,Y) = E \left[(X - \mu_x)^2\right] = Var(X) \ \ \ \ {\rm (for \ \ X=Y)}$$ 

Furthermore, if $\mu_x = \mu_y = 0$, then $Cov(X,Y) = E(XY)$.  Using the covariance, the correlation coefficient can be defined as

\begin{eqnarray*}
  r &=& \frac{Cov(X,Y)}{\sigma_x \sigma_y} \\[3mm]
    &=& \frac{E((X - \mu_x)(Y - \mu_y))}{\sigma_x \sigma_y} \\
\end{eqnarray*}

It is easy to show that the two definitions of $r$ are the same, i.e., if $Y = mX + Z$, then

\begin{eqnarray*}
  r &=& \frac{Cov(X,Y)}{\sigma_x \sigma_y} \\[3mm]
    &=& \frac{m \sigma_x}{\sigma_y} \\
\end{eqnarray*}

The advantage of the definition based on $Cov(X,Y)$ is that we do not need to know $m$. In most cases we are given a set of $n$ data points $(x_i, y_i)$ and we are required to find the correlation.  If we know the distribution of $X$ and $Y$ and hence $\sigma_x$ and $\sigma_y$, we can see effect of changing $r$.  This is shown in the figure below which is drawn using the the following R code.  Notice how $m$ and $\sigma_z$ are derived  from the given values of $r$, $\sigma_x$ and $\sigma_y$. You should be able to derive those relationships from the two definitions of $r$.



### Plots with different r values

```r
sigmax = 2    # assume sigma_x is known 
sigmay = 1    # assume sigma_y is known 
n = 1000

par(mfrow=c(2,3))     # set the plotting area to be 2 rows and 3 columns 
rvalues = c(0, 0.3, 0.5, 0.7, 0.9, 0.99)

for (r in rvalues){
  sigmaz = sigmay *  sqrt(1 - r**2)    
  m = r * sigmay/sigmax
  X = rnorm(n,1,sigmax)
  Z = rnorm(n,3,sigmaz)
  Y = m*X + Z
  plot(X,Y, main = paste("r = ", r))
}
```

## Correlation Coefficient Given Sampled Data
If we are given $n$ sampled data points $(x_i,y_i)$, the covariance is defined as

\begin{equation*}
  Cov(X,Y) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
\end{equation*}

where $\bar{x}$ and $\bar{y}$ are the sample means of the samples $x_i$ and $y_i$.

Furthermore, we can calculate the modified sample variances using

\begin{equation*}
  S_x^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
\end{equation*}

and

\begin{equation*}
  S_y^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - \bar{y})^2
\end{equation*}

Using the above we can now define $r$ as
\begin{equation*}
  r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}
\end{equation*}

###  Notes

Couple of notes

1. $S_x$ and $S_y$ and not unbiased estimators of the standard deviations of $X$ and $Y$ and neither is the $Cov(X,Y)$.


2. If we used unbiased estimator, then it should be divided by $n-1$. However, the final equation for $r$ will remain the same.


3. Given a set of data points the following are the steps to find the correlation coefficient $r$:  

a) Find $\bar{x}$ and $\bar{y}$. 

b) Find $S_x$ and $S_y$.

## The best-fit line
Give a set of points we want to draw the "best-fit" line. We define the best fit line to be the one that minimizes the sum of the squares of the vertical distances between the line and the points.  Suppose the line is given by $Y = Ax + B$ and we are given $n$ data points $(x_i,y_i)$. Then $\Gamma$, the sum of the squares of the vertical distances between the line and the points is given by

\begin{equation*}
  \Gamma = \sum_{i=1}^{n} (y_i - (Ax_i + B))^2
\end{equation*}
If we expand this and find $A$ and $B$ that minimizes $\Gamma$, we can show that $A = r \frac{S_y}{S_x}$.


## R Functions

1. cor :  computes the  correlation of x and y if these are vectors

2. cov :  computes the  covariance of x and y if these are vectors

3. lm  :  used to fit linear models

### Example

*** About the data set:***  These are the ages (in years) and the logarithmically transformed admission serum urea (natural logarithm [ln] urea) for 20 patients attending an accident and emergency unit (A\&E).



```r
data = read.table(file="./Data/urea-serum-data.txt", header=TRUE)
age = data[,2]
lnurea = data[,3]
plot(age, lnurea)
cor(age,lnurea)
lm(lnurea ~ age)
abline(coef=c(0.7147, 0.01716))
```

### Example 2

This example shows how to calculate $r$ from a data set using the steps discussed above. We also calculate $m$. We then use $lm$ to find the best fit line and show that the slope of the best fit line is the same as $m$. 

1. Create a data file with the data 


2. When you call lm, note that order. In  $lm(y \sim x)$, $y$ is the dependent variable or target (e.g., the admission serum urea) and $x$ is the independent variable or regressor (e.g., the age of the patient). $lm(x \sim y)$ is different. 

```r
data = read.table(file="./Data/lm_synthetic_data.txt", sep=",", header = FALSE)
x = data[,1]
y = data[,2]

barx = sum(x)/length(x)   # calculate sample mean for both X and Y
bary = sum(y)/length(y)

sx = sqrt(sum((x - barx)*(x - barx))/(length(x) - 1))  #calculate the Sample Variance for X and Y
sy = sqrt(sum((y - bary)*(y - bary))/(length(y) - 1))

covxy = sum((x - barx)*(y - bary))/(length(y) - 1)
r = covxy/(sx*sy)
print(r)
m = r * sy/sx
print(m)
plot(x,y)

lm_summary = lm(y ~ x)           # us lm to solve for the best fit line
intercept = coef(lm_summary)[1]
slope = coef(lm_summary)[2]
print(slope)
abline(coef=c(intercept, slope))
```
